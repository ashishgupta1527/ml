```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris.data.shape
```

```python
iris.target.shape
```

```python
X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=10)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: {:.2f}".format(mse))
print("Root mean squared error: {:.2f}".format(rmse))
print("R-squared value: {:.2f}".format(r2))
```

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_pca, y, test_size=0.33, random_state=10)
```

```python
X_train_2.shape
```

```python
lr.fit(X_train_2, y_train_2)
```

```python
y_pred_2 = lr.predict(X_test_2)
```
